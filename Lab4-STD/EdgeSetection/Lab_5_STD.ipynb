```python
# 1. Built-in Sobel, Prewill, Roberts

img = rgb2gray(io.imread('circuit.tif'))


img=img/255


img_prewitt = prewitt(img)
img_roberts = roberts(img)
img_sobel = sobel(img)
img_sobel_h =sobel_h(img)
img_sobel_v =sobel_v(img)
img_canny = canny(img,sigma=1.5,high_threshold=0.3,low_threshold=0.05)


show_images([img,img_prewitt,img_roberts,img_canny,img_sobel,img_sobel_h,img_sobel_v],["input","prewitt","roberts","canny","sobel","sobel_h","sobel_v"])

```

```python
# 2. Custom Sobel (V-H-Both)


Hx = np.array([[-1,0,1],
                [-2,0,2],
              [-1,0,1]])


Hy = np.array([[-1,-2,-1],
                [0,0,0],
              [1,2,1]])



img_x = convolve2d(img,Hx)
img_y = convolve2d(img,Hy)


img_x_p = np.multiply(img_x,img_x)
img_y_p = np.multiply(img_y,img_y)


sobel= np.sqrt(img_x_p+img_y_p)


#thresholding
sobel[sobel>=0.35]=1
sobel[sobel<0.35]=0


show_images([img,sobel,img_x,img_y],["input","sobel","sobel_x","sobel_y"])
```
